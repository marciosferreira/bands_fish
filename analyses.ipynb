```python
concatenated = concatenated.loc[concatenated['x_position'] != 'None']
concatenated['x_position'] = pd.to_numeric(concatenated["x_position"])
concatenated['direction'] = concatenated.apply(lambda row: "out", axis=1)
concatenated.loc[((concatenated['frame'] > r1) & (concatenated['frame'] < r2)), 'direction'] = 'right'
concatenated.loc[((concatenated['frame'] > l1) & (concatenated['frame'] < l2)), 'direction'] = 'left'
#test = concatenated.loc[concatenated['frame'] > 400]




```

```python
concatenated = concatenated.loc[concatenated['direction'] != 'out']
filtered_position_l = concatenated.loc[((concatenated['direction'] == "left") & (concatenated['x_position'] > 2624*0.10))]
filtered_position_r = concatenated.loc[((concatenated['direction'] == "right") & (concatenated['x_position'] < 2624*0.90))]
#filtered_position_n = concatenated.loc[((concatenated['direction'] == "stopped") & (concatenated['x_position'] < 2624*0.95) & (concatenated['x_position'] > 2624*0.05))]
filtered_l_by_id = filtered_position_l.groupby(['fish_id'])
filtered_r_by_id = filtered_position_r.groupby(['fish_id'])
##################
total_r = filtered_r_by_id['fish_id'].unique().count()
total_l = filtered_l_by_id['fish_id'].unique().count()
```

```python
filtered_position_l = filtered_position_l.loc[((filtered_position_l['value'] == 0) | (filtered_position_l['value'] < 0))]
filtered_position_r = filtered_position_r.loc[((filtered_position_r['value'] == 0) | (filtered_position_r['value'] > 0))]
```

```python
lft_group = filtered_position_l.groupby(['fish_id'])
lft_count = lft_group['value'].count()
lft_avg = lft_group['value'].mean()
lft_avg_final = lft_avg[((lft_avg > velocity*0.25) & (lft_avg < velocity*1.25))].count()
lft_avg_final

```

```python
right_group = filtered_position_r.groupby(['fish_id'])
rgt_count = right_group['value'].count()
rgt_avg = right_group['value'].mean()
rgt_avg_final = rgt_avg[((rgt_avg > velocity*0.25) & (rgt_avg < velocity*1.25))].count() 

rgt_avg_final
```

```python
abs_total = total_l + total_r
sum_result = lft_avg_final + rgt_avg_final
final_result = round(sum_result/abs_total*100, 2)
final_result
```
