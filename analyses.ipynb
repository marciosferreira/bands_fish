```python
names = list(df.columns)
filtered_names = [x for x in names if "frame_number" not in x]
filtered_names_again = [x for x in filtered_names if "pos" not in x]
filtered_names_again
```

```python

filtered_df = df[filtered_names_again]
filtered_df.reset_index(inplace=True)

filtered_df = filtered_df.rename(columns={"index": "frame"})
filtered_df
```

```python
names = list(filtered_df.columns)
filtered_names = [x for x in names if "frame" not in x]
print(filtered_names)

for idx, i in enumerate(filtered_names):
    new_df = filtered_df[['frame', i]]
    new_df['fish_id'] = i
    new_df.rename(columns={i: 'value'}, inplace=True)    
    if idx > 0:        
        concatenated = new_df.append(previous_df, ignore_index=True)        
        previous_df = concatenated.copy()
    else:
        previous_df = new_df.copy()
        
```

```python
concatenated        
```

```python
concatenated['direction'] = concatenated.apply(lambda row: "right", axis=1)
concatenated.loc[concatenated['value'] > 0, 'direction'] = 'right'
concatenated.loc[concatenated['value'] < 0, 'direction'] = 'left'
concatenated.loc[concatenated['value'] == 0, 'direction'] = 'stopped'
concatenated['abs'] = concatenated['value'].abs()
concatenated = concatenated[concatenated['abs'] != 0]
concatenated_left = concatenated[concatenated['direction'] == "left"]
concatenated_right = concatenated[concatenated['direction'] == "right"]
grouped_left = concatenated_left.groupby(['fish_id']).sum()
grouped_left.reset_index(inplace=True)
grouped_left = grouped_left.rename(columns={"index": "fish_id"})
grouped_left['direction'] = 'left'
grouped_right = concatenated_right.groupby(['fish_id']).sum()
grouped_right.reset_index(inplace=True)
grouped_right = grouped_right.rename(columns={"index": "fish_id"})
grouped_right['direction'] = 'right'
both_df = grouped_left.append(grouped_right, ignore_index=True)    
```
