```python

abs_l = filtered_position_l['value'].abs()
filtered_position_l['abs'] = abs_l
grouped_prov_l = filtered_position_l.groupby(['fish_id'])
dist = grouped_prov_l['distance'].mean()
val = grouped_prov_l['abs'].mean()
left_calc = dist/val
for i in left_calc.index:
    filtered_position_l.loc[filtered_position_l['fish_id'] == i, 'time'] = left_calc.loc[i] 


```

```python
abs_r = filtered_position_r['value'].abs()
filtered_position_r['abs'] = abs_r
grouped_prov_r = filtered_position_r.groupby(['fish_id'])
dist = grouped_prov_r['distance'].mean()
val = grouped_prov_r['abs'].mean()
right_calc = dist/val
for i in right_calc.index:
    filtered_position_r.loc[filtered_position_r['fish_id'] == i, 'time'] = right_calc.loc[i] 



```

```python
filtered_position_l = filtered_position_l.loc[filtered_position_l['time'] < 500]
filtered_position_r = filtered_position_r.loc[filtered_position_r['time'] < 500]
```

```python
filtered_l_by_id = filtered_position_l.groupby(['fish_id'])
filtered_r_by_id = filtered_position_r.groupby(['fish_id'])

total_r = filtered_r_by_id['fish_id'].unique().count()
total_l = filtered_l_by_id['fish_id'].unique().count()

print(total_r)
print(total_l)

```

```python
filtered_position_l = filtered_position_l.loc[(filtered_position_l['value'] > 0)]
filtered_position_r = filtered_position_r.loc[(filtered_position_r['value'] > 0)]
```

```python
lft_group = filtered_position_l.groupby(['fish_id'])
#lft_count = lft_group['value'].count()
lft_avg = lft_group['value'].mean()
lft_avg_final = lft_avg[((lft_avg > velocity*0.75) & (lft_avg < velocity*1.25))].count()
lft_avg

```

```python
right_group = filtered_position_r.groupby(['fish_id'])
#rgt_count = right_group['value'].count()
rgt_avg = right_group['value'].mean()
rgt_avg_final = rgt_avg[((rgt_avg > velocity*0.70) & (rgt_avg < velocity*1.30))].count()

```

```python
#abs_total = total_l + total_r
final_result_left = lft_avg_final/total_l*100
final_result_right = rgt_avg_final/total_r*100
final_result = statistics.mean([final_result_left, final_result_right])
#final_result = final_result_right
```

```python
gp_right = filtered_position_r.groupby(['fish_id'])
gp_left = filtered_position_l.groupby(['fish_id'])


r_result = gp_right['value'].mean()
l_result = gp_left['value'].mean()








```

```python


f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8,8))

sns.barplot(x=r_result, y=r_result.index, ax=ax1)
ax1.set_title(path_file + '_right' + ' - expected velocity: ' + str(velocity) + ' - fitted fish: ' + str(rgt_avg_final))
sns.barplot(x=l_result, y=l_result.index, ax=ax2)
ax2.set_title(path_file + '_left' + ' - expected velocity: ' + str(velocity) + ' - fitted fish: ' + str(lft_avg_final))
plt.tight_layout()
ax1.set_xlabel('average velocity')
ax2.set_xlabel('average velocity')



ax1.axvline(x=velocity*0.75, color='blue', linestyle='--')
ax1.axvline(x=velocity, color='red', linestyle='--')
ax1.axvline(x=velocity*1.25, color='blue', linestyle='--')
ax2.axvline(x=velocity*0.75, color='blue', linestyle='--')
ax2.axvline(x=velocity, color='red', linestyle='--')
ax2.axvline(x=velocity*1.25, color='blue', linestyle='--')

plt.show()


```
